```python
#eliminating outliers using IQR
Q1=marketing_campaign_df['Income'].quantile(0.25)
Q3=marketing_campaign_df['Income'].quantile(0.75)

IQR=Q3-Q1

lower=Q1-1.5*IQR
upper=Q3+1.5*IQR

marketing_campaign_df=marketing_campaign_df[(marketing_campaign_df['Income']>=lower)&(marketing_campaign_df['Income']<=upper)]
sns.boxplot(marketing_campaign_df['Income'])
plt.show()
```

```python
#create derived age columns
from datetime import datetime
current_year = datetime.now().year
marketing_campaign_df['Age']=current_year-marketing_campaign_df['Year_Birth']
marketing_campaign_df
```

```python
#create derived children columns
marketing_campaign_df['Children']=marketing_campaign_df['Kidhome']+marketing_campaign_df['Teenhome']
marketing_campaign_df
```

```python
#Create total spend column
marketing_campaign_df['Total_Spend'] = (
    marketing_campaign_df['MntWines']+
    marketing_campaign_df['MntFruits']+
    marketing_campaign_df['MntMeatProducts']+
    marketing_campaign_df['MntFishProducts']+
    marketing_campaign_df['MntSweetProducts']+
    marketing_campaign_df['MntGoldProds']
)
marketing_campaign_df
```

```python
#creating derived total purchase column
marketing_campaign_df['Total_Purchases']=(
    marketing_campaign_df['NumDealsPurchases']+
    marketing_campaign_df['NumWebPurchases']+
    marketing_campaign_df['NumCatalogPurchases']+
    marketing_campaign_df['NumStorePurchases']
)
marketing_campaign_df
```

```python
marketing_campaign_df['Family_Size'] = (
    marketing_campaign_df['Marital_Status'].map({
        'Single':1,
        'Married':2,
        'Together':2,
        'Widow':1,
        'Divorced':1,
        'Alone':1,
        'Absurd':2,
        'YOLO': 1
    })
    + marketing_campaign_df['Children']
)
marketing_campaign_df
```

```python
#customer since column
marketing_campaign_df['Customer_since']=(pd.Timestamp.today()-marketing_campaign_df['Dt_Customer']).dt.days
marketing_campaign_df
```

```python
marketing_campaign_df.drop(['Year_Birth'],axis=1,inplace=True)
marketing_campaign_df
```
