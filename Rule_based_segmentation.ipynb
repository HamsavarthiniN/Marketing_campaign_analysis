```python
import pandas as pd

clean_df = pd.read_pickle(r"D:\DataScience\GUVI\Local\Marketing_campaign_analysis\cleaned_data.pkl")
clean_df
```

```python
high_spend_threshold = clean_df['Total_Spend'].quantile(0.90)
# High Income
clean_df['High_Income'] = clean_df['Income'].apply(lambda x: 'Yes' if x > 86000 else 'No')

# Young Customer (Age < 45)
clean_df['Young_Customer'] = clean_df['Age'].apply(lambda x: 'Yes' if x < 45 else 'No')

# Campaign Responder
campaigns = ['AcceptedCmp1','AcceptedCmp2', 'AcceptedCmp3','AcceptedCmp4', 'AcceptedCmp5','Response']
clean_df['Campaign_Responder'] = clean_df[campaigns].apply(lambda row: 'Yes' if row.sum()>0 else 'No', axis=1)

# High Web Engagement
clean_df['High_Web_Engagement'] = clean_df['NumWebVisitsMonth'].apply(lambda x: 'Yes' if x > 7 else 'No')

# Family Customer
clean_df['Family_Customer'] = clean_df['Children'].apply(lambda x: 'Yes' if x > 0 else 'No')

# High Spender
clean_df['High_Spender'] = clean_df['Total_Spend'].apply(lambda x: 'Yes' if x > high_spend_threshold else 'No')
clean_df
```

```python
clean_df.to_pickle('incl_segmentation.pkl')
```
