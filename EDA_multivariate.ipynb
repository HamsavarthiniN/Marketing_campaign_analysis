```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

clean_df = pd.read_pickle(r"D:\DataScience\GUVI\Local\Marketing_campaign_analysis\cleaned_data.pkl")
clean_df
```

# Bivariate/Multivariate analysis

## Response vs income

```python
campaigns = ['AcceptedCmp1','AcceptedCmp2', 'AcceptedCmp3','AcceptedCmp4', 'AcceptedCmp5','Response']

#splits the Income column into 5 groups (quintiles), each containing approximately 20% of the customers
income_bins = pd.qcut(clean_df['Income'], q=5)

heatmap_data = clean_df.groupby(income_bins)[campaigns].mean() * 100

sns.heatmap(heatmap_data, annot=True, fmt='.1f', cmap='YlGnBu')

plt.title('Campaign Acceptance Rate (%) by Income Group')
plt.show()
```

## Response vs Age

```python
campaigns = ['AcceptedCmp1','AcceptedCmp2', 'AcceptedCmp3','AcceptedCmp4', 'AcceptedCmp5','Response']

#splits the Age column into 5 groups (quintiles), each containing approximately 20% of the customers
Age_bins = pd.qcut(clean_df['Age'], q=5)

heatmap_data = clean_df.groupby(Age_bins)[campaigns].mean() * 100

sns.heatmap(heatmap_data, annot=True, fmt='.1f', cmap='YlGnBu')

plt.title('Campaign Acceptance Rate (%) by Age Group')
plt.show()
```

## Response vs Education

```python
campaigns = ['AcceptedCmp1','AcceptedCmp2', 'AcceptedCmp3','AcceptedCmp4', 'AcceptedCmp5','Response']

heatmap_data = clean_df.groupby('Education')[campaigns].mean() * 100

sns.heatmap(heatmap_data, annot=True, fmt='.1f', cmap='YlGnBu')

plt.title('Campaign Acceptance Rate (%) by Education')
plt.show()
```

## Response vs total spend

```python
campaigns = ['AcceptedCmp1','AcceptedCmp2', 'AcceptedCmp3','AcceptedCmp4', 'AcceptedCmp5','Response']

spent_on = ['MntWines','MntFruits','MntMeatProducts', 'MntFishProducts','MntSweetProducts', 'MntGoldProds']

#create dataframe because each iteration gives only a Series
heatmap_data = pd.DataFrame()

for camp in campaigns:
    heatmap_data[camp] = clean_df.groupby(camp)[spent_on].mean().loc[1]

sns.heatmap(heatmap_data, annot=True, fmt='.0f', cmap='YlGnBu')

plt.title('Average Spending of Customers Who Accepted Each Campaign')
plt.ylabel('Product Category')
plt.show()
```

## Response vs New Purchase

```python
campaigns = ['AcceptedCmp1','AcceptedCmp2', 'AcceptedCmp3','AcceptedCmp4', 'AcceptedCmp5','Response']

purchase = ['NumDealsPurchases','NumWebPurchases','NumCatalogPurchases', 'NumStorePurchases']

#create dataframe because each iteration gives only a Series
heatmap_data = pd.DataFrame()

for camp in campaigns:
    heatmap_data[camp] = clean_df.groupby(camp)[purchase].mean().loc[1]

sns.heatmap(heatmap_data, annot=True, fmt='.0f', cmap='YlGnBu')

plt.title('Average New Purchase Made by Customers Who Accepted Each Campaign')

plt.show()
```

```python
campaigns = ['AcceptedCmp1','AcceptedCmp2', 'AcceptedCmp3','AcceptedCmp4', 'AcceptedCmp5','Response']

heatmap_data = clean_df.groupby('Country')[campaigns].mean() * 100

sns.heatmap(heatmap_data, annot=True, fmt='.1f', cmap='YlGnBu')

plt.title('Campaign Acceptance Rate (%) by Country')
plt.show()
```
